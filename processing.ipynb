```python
import os
from glob import glob
import shutil
from pathlib import Path

def delete_dir_paths(dir_path, except_list):
    """
    지정된 디렉토리에서 except_list에 포함되지 않은 모든 폴더와 파일을 삭제합니다.

    Args:
        dir_path (str): 삭제할 디렉토리 경로
        except_list (list): 삭제되지 않을 폴더 이름 목록

    Returns:
        list: 삭제된 폴더 목록
    """

    except_list = set(except_list)  # 빠른 검색을 위해 집합으로 변환
    deleted_dirs = []

    for path in Path(dir_path).glob("*"):  # pathlib를 사용하여 경로 처리
        if path.is_dir() and path.name not in except_list:
            try:
                shutil.rmtree(path)  # 폴더와 모든 하위 파일 삭제
                deleted_dirs.append(path)
            except OSError as e:
                print(f"폴더 삭제 실패: {path} ({e})")
        else:
            try:
                os.remove(path)  # 파일 삭제
            except OSError as e:
                print(f"파일 삭제 실패: {path} ({e})")

    return deleted_dirs

# 예시 사용법
dir_path = "D:\\상품 이미지\\Training\\[원천]과자1"  # 삭제할 디렉토리 경로
except_list = lists  # 삭제되지 않을 폴더 이름 목록
deleted_dirs = delete_dir_paths(dir_path, except_list)

print(f"삭제된 폴더: {deleted_dirs}")
```
